```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
f = open('stommel_npart_scaling.log', 'r')
lines = f.readlines()

n = len(lines)

npart = np.zeros(n)
wct = np.zeros(n)
wct_exec = np.zeros(n)
for i in range(n):
    l = lines[i].split()
    npart[i] = int(l[6])
    wct[i] = float(l[9])
#     wct_exec[i] = float(l[7][1:])

# args = np.argsort(nproc)
# nproc = nproc[args]
# wct = wct[args]


xlin_scaling = np.array([1e3,1e6])

fig = plt.figure(figsize=(7, 4.25), dpi=300, facecolor='w', edgecolor='k')
plt.loglog(xlin_scaling, xlin_scaling/55);

plt.xlabel('# particles');
plt.ylabel('WCT (s)');
plt.loglog(npart, wct, 'r');
plt.scatter(npart, wct, c='r');

plt.legend(['1:1', 'Stommel']);


plt.savefig('stommel_scaling_n_particles.png');

print("Time for one particle: %f" % wct[0])
```
